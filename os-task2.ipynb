# 1. Mean / Median / Mode Calculation

```python

def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 1:
        return sorted_numbers[n // 2]
    else:
        return (sorted_numbers[n//2 - 1] + sorted_numbers[n//2]) / 2

def calculate_mode(numbers):
    freq = {}
    for num in numbers:
        freq[num] = freq.get(num, 0) + 1
    max_count = max(freq.values())
    modes = [num for num, count in freq.items() if count == max_count]
    return modes[0] if len(modes) == 1 else modes
```


# 2. Variance Calculation

```python
def calculate_variance(numbers):
    m = calculate_mean(numbers)
    return sum((x - m) ** 2 for x in numbers) / len(numbers)
```

# 3. Standard Deviation Calculation

```python

def calculate_std_deviation(numbers):
    return calculate_variance(numbers) ** 0.5
```

# 4. Euclidean Distance Calculation

```python

def euclidean_distance(a, b):
    # If both are lists of the same length, calculate vector distance.
    if isinstance(a, list) and isinstance(b, list):
        if len(a) != len(b):
            return None  # Cannot compute distance if lengths differ.
        return sum((x - y) ** 2 for x, y in zip(a, b)) ** 0.5
    # Otherwise, treat them as single values.
    return abs(a - b)
```

# 5. Sigmoid Function

```python

def sigmoid(x):
    return 1 / (1 + 2.71828 ** (-x))

# --- Sample Usage ---
numbers = [1, 2, 2, 3, 4]
print("Numbers:", numbers)
print("Mean:", calculate_mean(numbers))
print("Median:", calculate_median(numbers))
print("Mode:", calculate_mode(numbers))
print("Variance:", calculate_variance(numbers))
print("Standard Deviation:", calculate_std_deviation(numbers))

print("\nEuclidean distance between 3 and 7:", euclidean_distance(3, 7))
print("Euclidean distance between [1,2,3] and [4,5,6]:", euclidean_distance([1,2,3], [4,5,6]))

print("\nSigmoid(0):", sigmoid(0))
print("Sigmoid(2):", sigmoid(2))
```
